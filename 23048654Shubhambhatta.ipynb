### Importing pandas and numpy as np, pd

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("Customer Service_Requests_from_2010_to_Present.csv")
df
```

### Converting created date and close date to datetime datatype 

```python
df['Created Date']=pd.to_datetime(df['Created Date'])
df['Closed Date']=pd.to_datetime(df['Closed Date'])
```

### Creating new columm request_closing_time

```python
df['Request_closing_time']=df['Closed Date']-df['Created Date']
```

```python
df['Day of Week'] = df['Created Date'].dt.day_name()
```

```python
complaints_by_day = df['Day of Week'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)
```

```python
plt.figure(figsize=(10, 6))
complaints_by_day.plot(kind='bar', color='skyblue')
plt.title('Number of Complaints by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_resolution = df.groupby('Borough')['Request_closing_time'].mean()

plt.figure(figsize=(8, 5))
plt.bar(avg_resolution.index, avg_resolution.dt.total_seconds() / 3600)
plt.ylabel('Avg Resolution Time (Hours)')
plt.title('Average Complaint Resolution Time by Borough')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
year_counts = df['Created Date'].dt.month.value_counts().sort_index()

plt.figure(figsize=(6, 6))
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%')
plt.title('Complaint Volume by Year')
plt.show()
```

```python
day_counts = df['Created Date'].dt.day_name().value_counts().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])
plt.figure(figsize=(10, 5))
plt.plot(day_counts.index, day_counts.values)
plt.title('Complaint Volume by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Complaints')
plt.tight_layout()
plt.show()
```

#### Arrange the complaint types according to their average 'Request_Closing_Time', categorized by various locations. Illustrate it through graph as well.

```python

df['Hours'] = df['Request_closing_time'].dt.total_seconds() / 3600
grouped = df.groupby(['Location Type', 'Complaint Type'])['Hours'].mean().reset_index()
pivot_df = grouped.pivot(index='Complaint Type', columns='Location Type', values='Hours')
pivot_df.plot(kind='bar', figsize=(14, 6))
plt.title('Avg Request Closing Time by Complaint Type and Location Type')
plt.xlabel('Complaint Type')
plt.ylabel('Avg Closing Time (Hours)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Test-1

```python
from scipy.stats import f_oneway
groups = [group['Hours'].values for name, group in df.groupby('Complaint Type')]
f_stat, p_value = f_oneway(*groups)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4e}")
```

### Test-2

```python
import pandas as pd
from scipy.stats import chi2_contingency

df = df.dropna(subset=['Complaint Type', 'Location Type'])
table = pd.crosstab(df['Complaint Type'], df['Location Type'])
chi2, p, dof, expected = chi2_contingency(table)

print(f"Chi-Square Statistic: {chi2:.3f}")
print(f"P-value: {p:.5f}")
print(f"Degrees of Freedom: {dof}")
```
